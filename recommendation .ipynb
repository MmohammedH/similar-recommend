```python
def extract_feature(img_path,model):
    img = cv2.imread(img_path)
    img=cv2.resize(img,(224,224))
    img =np.array(img)
    expand_img=np.expand_dims(img,axis=0)
    pre_image=preprocess_input(expand_img)
    result=model.predict(pre_image).flatten()
    normalized=result/norm(result)
    return normalized
```

```python
extract_feature('1550.jpg',model)
```

```python
class Solution:
    def getWinner(self, arr, k: int) -> int:
        count=0
        current=0
        while count<k:
            if arr[0]>arr[1]:
                current=arr[0]
                ele=arr.pop(1)
                arr.append(ele)
                if current==arr[0]:
                    count+=1
            else:
                current=arr[1]
                ele=arr.pop(0)
                arr.append(ele)
                if current==arr[1]:
                    count+=1
        return arr[0]
```

```python
# arr = [2,1,3,5,4,6,7]
# k = 2
arr=[3,2,1]
k = 10
obj=Solution()
obj.getWinner(arr,k)
```

```python
class Solution:
    def isPathCrossing(self, path: str) -> bool:

        origin=[0,0]
        for i in path:
            if i=='N':
                origin[0]+=1
            elif i=='S':
                origin[0]-=1
            elif i=='E':
                origin[1]+=1
            elif i=='W':
                origin[1]-=1
            if origin==[0,0]:
                return True
            print(origin)
        else:
            return False

```
